min 25 50 75 % max -
> 25% — ты прошёл четверть очереди. Все матчи до этой точки имеют ≤ этого значения.
Пример: 25% = 1 гол. Значит в каждом четвёртом матче хозяева забивали 0 или 1 гол.
50% — середина очереди. Половина матчей ниже, половина выше. Это медиана.
Пример: 50% = 1 гол. Значит в половине всех матчей хозяева забили 0-1 гол.
75% — прошёл три четверти. Только 25% матчей имеют больше этого значения.
Пример: 75% = 2 гола. Значит только каждый четвёртый матч заканчивался 3+ голами хозяев.
max — последний в очереди. Самый экстремальный результат в датасете.
